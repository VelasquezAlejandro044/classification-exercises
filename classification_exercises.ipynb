## 2.4
Create a function named `prep_titanic` that accepts the raw titanic data, and returns the data with the transformations above applied.

```python
def prep_titanic():
    titani_df = acquire.get_titanic_data()
    titani_df = titani_df.drop_duplicates()
    cols_to_drop = ['deck', 'embarked', 'class', 'age']
    titani_df = titani_df.drop(columns=cols_to_drop)
    titani_df['embark_town'] = titani_df.embark_town.fillna(value='Southampton')
    dummy_df = pd.get_dummies(titani_df[['sex','embark_town']], dummy_na=False, drop_first=[True, True])
    titani_df = pd.concat([titani_df, dummy_df], axis=1)
    return titani_df.sample(5)
    
```
